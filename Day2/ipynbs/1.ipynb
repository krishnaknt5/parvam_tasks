```python
# build a simple text analyzer using need libraries in python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from collections import Counter

# Download required datasets (only if not already installed)
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')

try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Sample Text
text = input("Enter your text:\n")

# Tokenization
sentences = sent_tokenize(text)
words = word_tokenize(text)

# Lowercase words
words_lower = [word.lower() for word in words if word.isalpha()]

# Remove stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words_lower if word not in stop_words]

# Word Frequency
word_freq = Counter(filtered_words)

# Most common words
common_words = word_freq.most_common(5)

# Simple Sentiment (basic approach)
positive_words = {"good", "great", "happy", "excellent", "love", "awesome"}
negative_words = {"bad", "sad", "hate", "terrible", "worst", "angry"}

pos_count = sum(1 for word in filtered_words if word in positive_words)
neg_count = sum(1 for word in filtered_words if word in negative_words)

# Results
print("\n--- TEXT ANALYSIS ---")
print("Total Sentences:", len(sentences))
print("Total Words:", len(words))
print("Filtered Words:", filtered_words)
print("Top 5 Words:", common_words)

print("\n--- SENTIMENT ---")
if pos_count > neg_count:
    print("Overall Sentiment: Positive 😊")
elif neg_count > pos_count:
    print("Overall Sentiment: Negative 😞")
else:
    print("Overall Sentiment: Neutral 😐")
```
