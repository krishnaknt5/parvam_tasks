**write a function that checks if two sentences are similar by comparing there word sets**

```python
def check_sentence_similarity(sentence1, sentence2):
    """
    Checks the similarity between two sentences by comparing their word sets.
    Returns a score between 0.0 and 1.0.
    """
    words1 = set(sentence1.lower().split())
    words2 = set(sentence2.lower().split())
    intersection = words1.intersection(words2)
    union = words1.union(words2)
    if len(union) == 0:
        return 0.0
    similarity = len(intersection) / len(union)
    return similarity
s1 = "The quick brown fox jumps over the lazy dog"
s2 = "A fast brown fox leaps over a lazy dog"

score = check_sentence_similarity(s1, s2)
print(f"Similarity Score: {score:.2f}")
if score > 0.5:
    print("Sentences are similar.")
else:
    print("Sentences are not similar.")
```
