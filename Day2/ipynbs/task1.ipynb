**write a function that tokenize a sentence into words and remove punctuations**

```python
import string

def tokenize_and_clean(sentence):
    table = str.maketrans('', '', string.punctuation)
    return [w.translate(table) for w in sentence.split() if w.translate(table)]

# Example
text = "Wait, is this working? Yes, it is!"
print(tokenize_and_clean(text))
```
